```python
def count_target_rows(workbook, length: int):
    """
    추출 대상인 노란 색상으로 칠해진 셀의 개수 체크

    Args:
        workbook (workbook): openpyxl로 연 workbook
        length (int): 전체 DataFrame의 길이

    Returns:
        color_cnt (int): 노란 색상의 개수
    """        
    ## 색상 기준 추출 대상 row 찾기 ##
    ws = workbook.get_sheet_by_name(workbook.get_sheet_names()[0])
    color_cnt: int = 0
    for row in range(2, length+1):
        try:
            if ws.cell(column=2, row=row).fill.fgColor.index == 43:  # ws.cell(column=1, row=row).fill.bgColor != '000000' and 
                color_cnt += 1
            
        except: pass
    return color_cnt

def count_target_rows2(workbook, length: int):
    """
    추출 대상인 노란 색상으로 칠해진 셀의 개수 체크

    Args:
        workbook (workbook): openpyxl로 연 workbook
        length (int): 전체 DataFrame의 길이

    Returns:
        color_cnt (int): 노란 색상의 개수
    """        
    ## 색상 기준 추출 대상 row 찾기 ##
    ws = workbook.get_sheet_by_name(workbook.get_sheet_names()[0])
    pre_color_index = '00000000'
    color_cnt: int = 0
    for row in range(2, length-1):
        try:
            now_color = ws.cell(column=2, row=row).fill.fgColor.index
            if now_color == 43:  # ws.cell(column=1, row=row).fill.bgColor != '000000' and 
                color_cnt += 1
                pre_color_index = ws.cell(column=2, row=row).fill.fgColor.index
            elif now_color == '00000000' and pre_color_index == 43:
                print(row, now_color)
                color_cnt += 1
        except: pass
    return color_cnt
```
